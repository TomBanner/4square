```python
import pandas as pd
import numpy as np
```

```python
offices_dataframe = pd.read_csv("officesproject.csv")
```

```python
offices_dataframe.head()
```

```python
offices_dataframe = offices_dataframe.drop(['Unnamed: 0'], axis=1)
```

```python
offices_dataframe = offices_dataframe.rename(columns={'Name' : 'Office Name'})
```

```python
for col in offices_dataframe:
        print(col)
```

```python
offices_dataframe.head()
```

```python
# create a list of categories
categories = ['International Airport', 'Train station', 'Nightclub', 'Pet grooming', 'Vegan restaurant', 
              'Basketball stadium', 'Design studio', 'Daycare', 'Preschool', 'High school']

# iterate over the categories
for category in categories:
    # create the new column names
    name_col = f"{category}: Name"
    dist_col = f"{category}: Distance"
    points_col = f"{category}: Points"
    # create a copy of the distance column and replace NaN and 0 values with NaN
    dist_col_copy = offices_dataframe[dist_col].copy()
    dist_col_copy.loc[(dist_col_copy == 0) | (dist_col_copy.isna())] = np.nan
    # create a new column with the points
    points = dist_col_copy.rank(method='min', ascending=True)
    points = points.apply(lambda x: 11 - x if 0 < x <= 10 else 0)
    # replace the NaN values in the points column with 0
    points[pd.isnull(points)] = 0
    # add the points column to the dataframe
    offices_dataframe[points_col] = points
```

```python
offices_dataframe.tail()
```

```python
# create a dictionary with the multipliers
multipliers = {'International Airport': 0.23, 'Train station': 0.23, 'Nightclub': 1.0, 'Pet grooming': 0.05, 
               'Vegan restaurant': 0.02, 'Basketball stadium': 0.02, 'Design studio': 0.23, 'Daycare': 0.33, 
               'Preschool': 0.33, 'High school': 0.33}

# loop through the categories and add the multiplier column
for category in multipliers.keys():
    points_column = f"{category}: Points"
    multiplier_column = f"{category}: Multiplier"
    offices_dataframe[multiplier_column] = multipliers[category]
```

```python
offices_dataframe.head()
```

```python
for col in offices_dataframe.columns:
    if "Points" in col:
        category = col.split(":")[0]
        multiplier_col = category + ": Multiplier"
        weighted_col = category + ": Weighted Score"
        offices_dataframe[weighted_col] = offices_dataframe[col] * offices_dataframe[multiplier_col]
```

```python
offices_dataframe.tail()
```

```python
weighted_cols = [col for col in offices_dataframe.columns if "Weighted Score" in col]
offices_dataframe["Final Score"] = offices_dataframe[weighted_cols].sum(axis=1)
```

```python
offices_dataframe.tail()
```

```python
highest_score = offices_dataframe.loc[offices_dataframe["Final Score"].idxmax()]
print(f"Location with highest score: {highest_score.name}")
```
